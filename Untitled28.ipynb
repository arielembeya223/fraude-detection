```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
import joblib
from datetime import datetime

def save_model(model, scaler):
    pipeline = Pipeline([
        ('scaler', scaler),
        ('model', model)
    ])
    joblib.dump(pipeline, 'fraud_detection_fin.pkl')
    print("Modèle et pipeline sauvegardés avec succès")

def load_balanced_data(sample_frac=0.5):
    trans = pd.read_csv('')#train transaction
    ident = pd.read_csv('')#train identity
    data = pd.merge(trans, ident, on='TransactionID', how='left')

    fraud = data[data['isFraud'] == 1]
    non_fraud = data[data['isFraud'] == 0]

    non_fraud_sampled = non_fraud.sample(n=len(fraud), random_state=42)  # Ratio 1:1
    balanced_data = pd.concat([fraud, non_fraud_sampled]).sample(frac=sample_frac, random_state=42)

    return balanced_data

def preprocess_data(df):
    df = df.copy()
    df['TransactionDT'] = pd.to_datetime(df['TransactionDT'], unit='s', errors='coerce')
    cat_cols = df.select_dtypes(include=['object']).columns
    df[cat_cols] = df[cat_cols].fillna('unknown')
    num_cols = df.select_dtypes(include=np.number).columns
    df[num_cols] = df[num_cols].fillna(df[num_cols].median())
    for col in cat_cols:
        df[col] = pd.factorize(df[col])[0]
    return df

def build_transaction_graph(df, max_nodes=1000):
    G = nx.DiGraph()
    unique_clients = df['card1'].dropna().unique()
    if len(unique_clients) > max_nodes / 2:
        selected_clients = np.random.choice(unique_clients, size=int(max_nodes / 2), replace=False)
        df = df[df['card1'].isin(selected_clients)]

    for _, row in df.iterrows():
        client = row['card1']
        merchant = row['ProductCD']
        if not G.has_node(client):
            G.add_node(client, type='client', isFraud=row['isFraud'])
        if not G.has_node(merchant):
            G.add_node(merchant, type='merchant')
        G.add_edge(client, merchant,
                   amount=row['TransactionAmt'],
                   is_fraud=row.get('isFraud', np.nan),
                   time=row['TransactionDT'])
    return G

def extract_features(G):
    features = []
    labels = []
    nodes = list(G.nodes)
    for node in nodes:
        degree = G.degree(node)
        in_degree = G.in_degree(node)
        out_degree = G.out_degree(node)
        amounts = [data['amount'] for _, _, data in G.edges(node, data=True)]
        total_amount = sum(amounts)
        avg_amount = np.mean(amounts) if amounts else 0
        max_amount = max(amounts) if amounts else 0
        if G.nodes[node]['type'] == 'client':
            times = [data['time'] for _, _, data in G.out_edges(node, data=True)]
            if len(times) > 1:
                time_diffs = np.diff(sorted(times))
                avg_time_diff = np.mean([td.total_seconds() for td in time_diffs])
            else:
                avg_time_diff = 0
        else:
            avg_time_diff = 0
        features.append([degree, in_degree, out_degree, total_amount, avg_amount, max_amount, avg_time_diff])
        label = G.nodes[node].get('isFraud', 0)
        labels.append(label)
    return np.array(features), np.array(labels), nodes

def main():
    train = load_balanced_data()
    train = preprocess_data(train)
    G = build_transaction_graph(train, max_nodes=1000)
    print(f"Graphe créé avec {len(G.nodes)} nœuds et {len(G.edges)} arêtes")

    X, y_true, nodes = extract_features(G)
    mask_clients = [G.nodes[n]['type'] == 'client' for n in nodes]

    X_clients = X[mask_clients]
    y_clients = y_true[mask_clients]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_clients)

    # Modèle Random Forest
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_model.fit(X_scaled, y_clients)

    # Prédictions en probabilités
    y_proba = rf_model.predict_proba(X_scaled)[:, 1]
    y_pred = (y_proba >= 0.5).astype(int)  # Seuillage à 0.5

    acc = accuracy_score(y_clients, y_pred)
    prec = precision_score(y_clients, y_pred, zero_division=0)
    rec = recall_score(y_clients, y_pred, zero_division=0)
    f1 = f1_score(y_clients, y_pred, zero_division=0)
    auc = roc_auc_score(y_clients, y_proba)

    print("\n### Évaluation des performances - Random Forest ###")
    print(f"Accuracy  : {acc:.4f}")
    print(f"Precision : {prec:.4f}")
    print(f"Recall    : {rec:.4f}")
    print(f"F1-score  : {f1:.4f}")
    print(f"ROC AUC   : {auc:.4f}")
    print("Matrice de confusion - Random Forest:")
    print(confusion_matrix(y_clients, y_pred))

    save_model(rf_model, scaler)

    # Visualisation des anomalies
    plt.figure(figsize=(20, 15))
    pos = nx.spring_layout(G, k=0.15, iterations=50, seed=42)
    node_colors, node_sizes = [], []

    client_indices = np.where(mask_clients)[0]
    anomalies = np.where(y_pred == 1)[0]  # Indices anomalies détectées
    
    for i, node in enumerate(nodes):
        if G.nodes[node]['type'] == 'client' and i in client_indices:
            node_index = client_indices.tolist().index(i)
            node_colors.append('red' if y_pred[node_index] == 1 else 'green')
            node_sizes.append(100 if y_pred[node_index] == 1 else 30)
        else:
            node_colors.append('blue')
            node_sizes.append(20)

    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes, alpha=0.8)
    nx.draw_networkx_edges(G, pos, edge_color='gray', alpha=0.05, width=0.5)

    plt.title(f"Détection de Fraude - {len(G.nodes)} nœuds ({len(anomalies)} anomalies détectées)")
    plt.axis('off')
    plt.tight_layout()
    plt.show()

    return {nodes[client_indices[i]]: y_proba[i] for i in anomalies}  # On renvoie les probabilités des anomalies

if __name__ == '__main__':
    anomalies = main()
    print(f"\nNombre d'anomalies détectées : {len(anomalies)}")
    print("\nTop 10 anomalies (Node : Probabilité de fraude) :")
    for node, proba in sorted(anomalies.items(), key=lambda x: -x[1])[:10]:
        print(f"Node {node} : {proba:.4f}")
```
